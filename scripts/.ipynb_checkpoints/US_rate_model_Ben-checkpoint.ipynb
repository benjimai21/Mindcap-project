```python
def pick_var_interest(df, df_dict):
    
    predictors = df_dict[df_dict["USEFUL FOR STUDY?"] == "x"]["VARIABLE NAME"].to_list()
    
    predictors.append("INSTNM")
    
    df_X = df[predictors]
    name_y_var = df_dict[df_dict["USEFUL FOR STUDY?"] == "Y"]["VARIABLE NAME"].iloc[1]
    
    df_y = df[name_y_var]
    
    df_X[name_y_var] = df_y
    
    df_var = df_X
    
    #remove duplicate universities (they will have the same average default rate so it does not make sense to keep it)
    df_var = df_var.drop_duplicates(subset = ["INSTNM"])
    
    df_var = df_var.set_index("INSTNM")
    
    return df_var

df_var = pick_var_interest(df, df_dict)

```

```python
def deal_missing_values(df):
    
    df = df.dropna(subset = ["CDR3"])
    
    df = df.dropna(thresh = 5000, axis = 1)
    
    number_col = df.shape[1]
    
    df = df.dropna(thresh = number_col, axis = 0) 
    
    return df

df_var_missdone = deal_missing_values(df_var)
```

```python
def convert_stringcol_to_floatcol(df):
    
    df = df.apply(lambda x: pd.to_numeric(x,errors='coerce'))
    
    return df

df_clean = convert_stringcol_to_floatcol(df_var_missdone)
```

```python
def train_test_split_method(df, y_col, test_size = 0.15):
    '''
    Divides data into training and test set depending of the method of encoding
    
        @df_filter_students_with_income_clean_hot_method: Dataframe encoded with one hot encoding
        @df_filter_students_with_income_clean_cat_method: Dataframe encoded with cat encoding
        @hot: Boolean if hot encoding is used
    '''
    
    
    X = df.drop(y_col, axis = 1)
    y = df[y_col]

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = test_size, random_state = SEED)
    
    return X_train, X_test, y_train, y_test

```

```python
def get_features_importance(model, X):

    plt.figure()
    
    features_importance = pd.Series(model.feature_importances_, index = X.columns)

    sorted_importances_rf = features_importance.sort_values()[-15:]
  
    sorted_importances_rf.plot(kind = "barh", color = "purple")
    plt.show()
```

```python
def build_scatter_plot(df, x, y, x_label, y_label, title):

    fig, ax = plt.subplots()

    ax.scatter(df[x], df[y], alpha = 0.7, s = 2)
    
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    ax.set_title(title)
    
    plt.show()

l = ["PELL_EVER", "FAMINC", "DEBT_MDN"]

for var in l:
    
    build_scatter_plot(df_clean, var, "CDR3", var, "CDR3", "default rate to " + var)
```

```python
def build_bar_plot(col_name, bar_height, x_label, y_label, title):
    
    bar_positions = range(len(bar_height))
    tick_positions = range(len(bar_height))
    
 
    fig, ax = plt.subplots()
        
    ax.bar(bar_positions, bar_height, 0.5)
    ax.set_xticks(tick_positions)
    ax.set_xticklabels(col_name)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    ax.set_title(title)
    
    #file_path = "../figures/Interest_rate_dataset/" + title + ".png"
    
    #plt.savefig(file_path)
    plt.show()
    
corr_list = []
for var in l:
    
    corr_list.append(np.corrcoef(df_clean[var], df_clean["CDR3"])[1,0])

build_bar_plot(l, corr_list, "column names", "correlation", "correlation of most important variables")
    
```
